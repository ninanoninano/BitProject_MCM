```python
le = LabelEncoder()
df['REPORTED_USAGE_LEVEL'] = le.fit_transform(df['REPORTED_USAGE_LEVEL'])
```

```python
cols = df.select_dtypes('object').columns.to_list()  #원핫인코딩
```

```python
df = pd.get_dummies(data=df, columns=cols, drop_first=True)
```

```python
 target = 'CHURN' # 분할
 X = df.drop(target, axis=1)
 y = df[target]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2023)
X_train
```

```python
from tensorflow.keras.backend import clear_session
from tensorflow.keras.models import Sequential, load_model   #레이어(계층)
from tensorflow.keras.layers import Dense, Dropout, Input     #레이어 관련 함수
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint   
#콜백 관련 함수
from tensorflow.keras.utils import to_categorical
```

```python
X_train.shape
```

```python
clear_session()

model = Sequential([           
	Input(shape=(18,)),
	Dense(64, activation='relu'),
	Dropout(0.2),
	Dense(32, activation='relu'),
	Dropout(0.2),
	Dense(16, activation='relu'),
	Dropout(0.2),
	Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4, verbose=1)
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', verbose=1, save_best_only=True)

hist = model.fit(X_train, y_train, batch_size=32, epochs=100, callbacks=[es, mc], validation_split=(X_test, y_test), verbose=1)
```

```python
import matplotlib.pyplot as plt

plt.title('test')
plt.plot(hist.history['acc'], label='acc') # 출력 그래프 데이터
plt.plot(hist.history['val_acc'], label='val_acc')
plt.plot(hist.history['loss'], label='loss')
plt.plot(hist.history['val_loss'], label='val_loss')
plt.legend() #범례
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.show()
```
